## Summary of Outlier Analysis: 



We checked outliers for all of the variables. The ones we focused on were BMI–the only continuous variable–and the ordinal variables with the larget ranges, PhysHlth, MentHlth, and GenHlth.

We visualized the distributions of the above variables with violin plots and found skews and outliers in the graphs. However, we will not remove any outliers. The skew can be accounted for by the natural variation in the sample population. The more extreme points are considered true outliers since they do not follow the majority distribution but are still accurate. We will keep these true outliers, as they represent our sample population and provide information for accurate predictions.

We considered using DBSCAN to visualize the outliers. However, we only have one continuous variable, BMI, so the clusters would not be visualized well.

```python
from sklearn.cluster import DBSCAN
import pandas as pd
#% matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('diabetes_binary_health_indicators_BRFSS2015.csv')
df.head()
```

# Variables with Potential Outliers

Variable Explanations From Kaggle:
- MentHlth: Now thinking about your mental health, which includes stress, depression, and problems with emotions, for how many days during the past 30 days was your mental health not good? scale 1-30 days
- PhysHlth: Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30 days was your physical health not good? scale 1-30 days
- BMI: Body Mass Index 

```python
df["diabetes_binary"]= df["Diabetes_binary"].replace({0:"Non-Diabetic",1:"Diabetic"})
```

```python
sns.violinplot(x ='GenHlth', hue='diabetes_binary', data=df)
plt.title('General Health Rating Distribution')
plt.xlabel('General Health Rating')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
sns.violinplot(x ='MentHlth', hue='diabetes_binary', data=df)
plt.title('Mental Health Rating Distribution')
plt.xlabel('Mental Health Rating')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
sns.violinplot(x = 'PhysHlth', hue='diabetes_binary', data=df)
plt.title('Physical Health Rating Distribution')
plt.xlabel('Physical Health Rating')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
sns.violinplot(x = 'BMI', hue='diabetes_binary', data=df)
plt.title('BMI Distribution')
plt.xlabel('BMI Score')
plt.ylabel('Density')
plt.grid()
plt.show()
```
